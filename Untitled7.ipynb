## Import and upload the file

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['climate_change_impact_on_agriculture_2024.csv']))
```

```python
# Check missing values
print(df.isnull().sum())

# Fill missing values
df.fillna({
    'Average_Temperature_C': df['Average_Temperature_C'].median(),
    'Total_Precipitation_mm': df['Total_Precipitation_mm'].median(),
    'CO2_Emissions_MT': df['CO2_Emissions_MT'].median(),
    'Irrigation_Access_%': df['Irrigation_Access_%'].median(),
    'Pesticide_Use_KG_per_HA': df['Pesticide_Use_KG_per_HA'].median(),
    'Fertilizer_Use_KG_per_HA': df['Fertilizer_Use_KG_per_HA'].median(),
    'Soil_Health_Index': df['Soil_Health_Index'].median(),
}, inplace=True)

# Drop rows with remaining missing values (if any)
df.dropna(inplace=True)
```

# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crop Yield distribution
sns.histplot(df['Crop_Yield_MT_per_HA'], kde=True)
plt.title("Distribution of Crop Yield")
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Average yield by region
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Crop_Yield_MT_per_HA', data=df)
plt.xticks(rotation=90)
plt.title("Crop Yield by Region")
plt.show()
```

# Preprocessing

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# Encode categorical features
label_cols = ['Country', 'Region', 'Crop_Type', 'Adaptation_Strategies']
for col in label_cols:
    df[col] = LabelEncoder().fit_transform(df[col])

# Select features and target
X = df.drop(columns=['Crop_Yield_MT_per_HA'])
y = df['Crop_Yield_MT_per_HA']

# Normalize features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

# Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import lightgbm as lgb

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Random Forest
rf = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
rf.fit(X_train, y_train)

# LightGBM
lgb_model = lgb.LGBMRegressor(learning_rate=0.05, n_estimators=500)
lgb_model.fit(X_train, y_train, eval_set=[(X_test, y_test)])
```

# Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
import pandas as pd
from google.colab import files
import io


# Fill missing values
df.fillna({
    'Average_Temperature_C': df['Average_Temperature_C'].median(),
    'Total_Precipitation_mm': df['Total_Precipitation_mm'].median(),
    'CO2_Emissions_MT': df['CO2_Emissions_MT'].median(),
    'Irrigation_Access_%': df['Irrigation_Access_%'].median(),
    'Pesticide_Use_KG_per_HA': df['Pesticide_Use_KG_per_HA'].median(),
    'Fertilizer_Use_KG_per_HA': df['Fertilizer_Use_KG_per_HA'].median(),
    'Soil_Health_Index': df['Soil_Health_Index'].median(),
}, inplace=True)

# Drop rows with remaining missing values (if any)
df.dropna(inplace=True)

# Encode categorical features
label_cols = ['Country', 'Region', 'Crop_Type', 'Adaptation_Strategies']
for col in label_cols:
    df[col] = LabelEncoder().fit_transform(df[col])

# Select features and target
X = df.drop(columns=['Crop_Yield_MT_per_HA'])
y = df['Crop_Yield_MT_per_HA']

# Normalize features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


# Train Random Forest
rf = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
rf.fit(X_train, y_train)

# Train LightGBM
lgb_model = lgb.LGBMRegressor(learning_rate=0.05, n_estimators=500)
lgb_model.fit(X_train, y_train, eval_set=[(X_test, y_test)])


# Random Forest
rf_preds = rf.predict(X_test)
print("Random Forest R:", r2_score(y_test, rf_preds))
print("Random Forest RMSE:", np.sqrt(mean_squared_error(y_test, rf_preds)))

# LightGBM
lgb_preds = lgb_model.predict(X_test)
print("LightGBM R:", r2_score(y_test, lgb_preds))
print("LightGBM RMSE:", np.sqrt(mean_squared_error(y_test, lgb_preds)))
```

# PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualize PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=10)
plt.colorbar(label='Crop Yield')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('PCA of Climate-Agriculture Dataset')
plt.show()
```

## Hyperparameter Tuning

```python
# Step 1: Fit Random Forest
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
rf_model.fit(X_train, y_train)

# Step 2: Fit LightGBM
import lightgbm as lgb

lgb_model = lgb.LGBMRegressor(learning_rate=0.05, n_estimators=500)
lgb_model.fit(X_train, y_train)

# Step 3: Evaluate Models
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd

# Random Forest predictions
rf_preds = rf_model.predict(X_test)
rf_r2 = r2_score(y_test, rf_preds)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_preds))

# LightGBM predictions
lgb_preds = lgb_model.predict(X_test)
lgb_r2 = r2_score(y_test, lgb_preds)
lgb_rmse = np.sqrt(mean_squared_error(y_test, lgb_preds))

# Step 4: Print Evaluation Metrics
print(" Model Performance:")
print(f"Random Forest  R: {rf_r2:.2f}, RMSE: {rf_rmse:.3f}")
print(f"LightGBM       R: {lgb_r2:.2f}, RMSE: {lgb_rmse:.3f}")

# Step 5: Choose Best Model
if lgb_r2 > rf_r2:
    best_model_name = "LightGBM"
    best_preds = lgb_preds
else:
    best_model_name = "Random Forest"
    best_preds = rf_preds

print(f"\n Best Model Selected: {best_model_name}")

# Step 6: Display Predictions
predicted_df = pd.DataFrame({
    'Actual Crop Yield': y_test.values,
    'Predicted Crop Yield': best_preds
}).reset_index(drop=True)

print("\n Sample Predictions (Actual vs Predicted):")
print(predicted_df.head(10))
```
